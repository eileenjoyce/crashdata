```r
#Combine crash data years
total_crash = rbind(crash_2016,crash_2015,crash_2014,crash_2013,crash_2012,crash_2011,crash_2010,crash_2009,
                    crash_2008,crash_2007,crash_2006,crash_2005,crash_2004,crash_2003,crash_2002,crash_2001,crash_2000,
                    crash_1999,crash_1998,crash_1997)
head(total_crash)
```

```r
#combine motorcycle data years
total_cycle = rbind(abbv_cycle_2016,abbv_cycle_2015,abbv_cycle_2014,abbv_cycle_2013,abbv_cycle_2012,abbv_cycle_2011,abbv_cycle_2010,abbv_cycle_2009,
                    abbv_cycle_2008,abbv_cycle_2007,abbv_cycle_2006,abbv_cycle_2005,abbv_cycle_2004,abbv_cycle_2003,abbv_cycle_2002,abbv_cycle_2001,
                    abbv_cycle_2000,abbv_cycle_1999,abbv_cycle_1998,abbv_cycle_1997)
```

```r
#add helmet variables by crash number
total_crash$MC_PASSNGR_IND = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_PASSNGR_IND"] 
total_crash$MC_DVR_HLMTON_IND = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_DVR_HLMTON_IND"] 
total_crash$MC_DVR_HLMTDOT_IND = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_DVR_HLMTDOT_IND"] 
total_crash$MC_DVR_HLMT_TYPE = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_DVR_HLMT_TYPE"] 
total_crash$MC_PAS_HLMTON_IND = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_PAS_HLMTON_IND"] 
total_crash$MC_PAS_HLMTDOT_IND = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_PAS_HLMTDOT_IND"] 
total_crash$MC_PAS_HLMT_TYPE = total_cycle[match(total_crash$CRN, total_cycle$CRN),"MC_PAS_HLMT_TYPE"] 



head(total_crash)
summary(total_crash)
str(total_crash)
```

```r
#trim year off county year variable
total_crash$COUNTY_YEAR <- as.character(total_crash$COUNTY_YEAR)
total_crash$COUNTY_YEAR = substr(total_crash$COUNTY_YEAR,12,(length(total_crash$COUNTY_YEAR)-12))
```

```r
#mcycle crashes by county by year
state_crashes <- 
total_crash %>% 
group_by(CRASH_YEAR,COUNTY_YEAR) %>% 
summarize(total = sum(MOTORCYCLE_COUNT))%>%
arrange(COUNTY_YEAR,CRASH_YEAR,total)%>%
print(state_crashes, n = Inf)
```

```r
#mcycle fatals by county by year
state_fatals <- 
total_crash %>% 
group_by(COUNTY_YEAR,CRASH_YEAR) %>% 
summarize(total = sum(MCYCLE_DEATH_COUNT))%>%
arrange(COUNTY_YEAR,CRASH_YEAR,total)%>%
print(state_fatals, n = Inf)
```

```r
#total motorcycle fatals by year
state_fatals1 <- 
total_crash %>% 
group_by(CRASH_YEAR) %>%
summarize(total = sum(MCYCLE_DEATH_COUNT))%>%
arrange(CRASH_YEAR)
state_fatals1
```

```r
#total motocycle fatals since 2003 by county
state_fatal2 <- 
total_crash %>% 
filter(CRASH_YEAR >2003) %>%
group_by(COUNTY_YEAR) %>% 
summarize(total = sum(MCYCLE_DEATH_COUNT))%>%
arrange(total)%>%
print(state_fatal2, n = Inf)
```

```r
#total crashes since 2003 by county
state_crash2 <- 
total_crash %>% 
filter(CRASH_YEAR >2003) %>%
group_by(COUNTY_YEAR) %>% 
summarize(total = sum(MOTORCYCLE_COUNT))%>%
arrange(total)%>%
print(state_crash2, n = Inf)
```

```r
total_crash$no_helmet<- total_crash$MC_DVR_HLMTON_IND == "N"
```

```r
#helmetless crashes total by county
state_no <- 
total_crash %>% 
filter(!is.na(no_helmet) & no_helmet != "NULL") %>%
filter(no_helmet == "TRUE")%>%
group_by(COUNTY_YEAR,no_helmet)%>%
tally(sort = T) %>%
ungroup() %>%
arrange(desc(n))%>%
print(state_no, n = Inf)
```

```r
#helmetless crashes as proportion of total


final <-left_join(state_no,state_crash2)
#final

helmet_prop<-final %>%
group_by(COUNTY_YEAR,total,n)%>%
summarise(total/n)%>%
ungroup() %>%
arrange(desc(total/n))

helmet_prop

```

```r
helmet_prop<-final %>%
group_by(COUNTY_YEAR,total,n)%>%
summarise(total/n)%>%
ungroup() %>%
arrange(desc(total/n))
```

```r
#proportion of helmetless crashes to total fatals
final_death <-left_join(state_no,state_fatal2)
#final_death

final_death %>%
group_by(COUNTY_YEAR,total,n)%>%
summarise(total/n)%>%
ungroup() %>%
arrange(desc(total/n))%>%
print(helmet_prop, n = Inf)
```
